```python
import numpy as np

import matplotlib.pyplot as plt

def equation1(x:float)->float:
    return ((x**2) - 1)

def equation2(x:float)->float:
    return np.exp(1-x**2)

# Generar los valores de X
x = np.linspace(-5, 5, 200)

# Calcular los valores de Y
y1 = equation1(x)
y2 = equation2(x)

#Ponerle color a las gráficas
plt.plot(x, y1, color='purple', label='$x^{2}-1$')  
plt.plot(x, y2, color='green', label='$e^{1-x^{2}}$')  

plt.xlabel('x')
plt.ylabel('y')
plt.title('Trama de $x^{2}-1$ y $e^{1-x^{2}}$')
ax = plt.gca()
ax.set_ylim([-2, 5])
ax.set_xlim([-4, 4])
plt.grid(True)
plt.legend()

# Marca los ejes coordenados
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
def eqBis(x):
    g = (x**2) - 1
    h = np.exp(1-x**2)
    return g-h

# Genera los valores de x
x = np.linspace(-5, 5, 100)
# Calcula los valores de y 
y = eqBis(x)
# Trama de la Ecuación
plt.plot(x, y, label = '$w(x)-z(x)$', color = 'cyan')
#Etiqueta los ejes
plt.xlabel('x')
plt.ylabel('y')
#Titulo de la gráfica
plt.title('Trama de $f(x)= w(x)-z(x)$')
#Configura los límites de los ejes
ax = plt.gca()
ax.set_ylim([-2, 5])
ax.set_xlim([-4, 4])
#añade una cuadrícula a la gráfica
plt.grid(True)
#Añade la leyenda
plt.legend()
# Marca los ejes coordenados
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
#Muestra la gráfica
plt.show()
```
